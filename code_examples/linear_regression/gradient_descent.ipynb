```python
def func_grad(x: float) -> float:
  return 2 * x

x = 1
lst_x = []
lst_x.append(x)

alpha = 0.1
err = 1e-6
grad = func_grad(x)

iter = 0
```

```python
while abs(grad) > err:
  grad = func_grad(x)
  x = x - alpha * grad
  lst_x.append(x)
  iter += 1
```

```python
print(f"Iterations = {iter}")
print(f"x(min) = {x}")
print(grad)
```
