```python
def area(a: int, b: int) -> int:
    return a*b
```

```python
print("Площадь прямоугольника: ",int(input("Введите длину стороны: a"))*int(input("Введите длину стороны: b")))
```

```python
class Area:
    def __init__(self,side_a,side_b):
        self.side_a=side_a
        self.side_b=side_b

    def calc(self):
        ab=self.side_a * self.side_b
        return f"Площадь прямоугольника со сторонами {self.side_a} и {self.side_b} равна {ab}."
```

```python
area=Area(int(input("Введите длину стороны: a")),int(input("Введите длину стороны: b")))
print(area.calc())
```
